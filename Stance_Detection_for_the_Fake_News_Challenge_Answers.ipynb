#### fit_on_texts() gives the following attributes in the output as given [here](https://faroit.github.io/keras-docs/1.2.2/preprocessing/text/).

* **word_counts:** dictionary mapping words (str) to the number of times they appeared on during fit. Only set after fit_on_texts was called.

* **word_docs:** dictionary mapping words (str) to the number of documents/texts they appeared on during fit. Only set after fit_on_texts was called.

* **word_index:** dictionary mapping words (str) to their rank/index (int). Only set after fit_on_texts was called.

* **document_count:** int. Number of documents (texts/sequences) the tokenizer was trained on. Only set after fit_on_texts or fit_on_sequences was called.

### Create embedding matrix with the glove embeddings


Run the below code to create embedding_matrix which has all the words and their glove embedding if present in glove word list.

````python
# load the whole embedding into memory
embeddings_index = dict()
f = open('./glove.6B.100d.txt')
for line in f:
	values = line.split()
	word = values[0]
	coefs = np.asarray(values[1:], dtype='float32')
	embeddings_index[word] = coefs
f.close()
print('Loaded %s word vectors.' % len(embeddings_index))

# create a weight matrix for words in training docs
#embedding_matrix = np.zeros((vocab_size, 100))
embedding_matrix = np.zeros((len(t.word_index)+1, 100))

for word, i in t.word_index.items():
	embedding_vector = embeddings_index.get(word)
	if embedding_vector is not None:
		embedding_matrix[i] = embedding_vector
````

# Try the sequential model approach and report the accuracy score. [10 marks]  

### Import layers from Keras to build the model

````python
from keras.models import Sequential,Model
from keras.layers import Embedding,Dense,LSTM,Dropout,Flatten,Input,Bidirectional,concatenate,dot
from keras.optimizers import SGD
````

### Model

````python
head_input = Input(shape=(400,), dtype='int32', name='head_input')
body_input = Input(shape=(400,), dtype='int32', name='body_input')
shared_embed = Embedding(len(t.word_index) + 1,100,weights=[embedding_matrix],trainable=False)
head_embed = shared_embed(head_input)
body_embed = shared_embed(body_input)

head_lstm = Bidirectional(LSTM(64,dropout=0.2, recurrent_dropout=0.2, name='head_lstm'))(head_embed)
body_lstm = Bidirectional(LSTM(64,dropout=0.2, recurrent_dropout=0.2, name='body_lstm'))(body_embed)
conc = concatenate([head_lstm,body_lstm])
#dense = Dense(100,activation='relu')(conc)
dense = Dense(4,activation='softmax')(conc)
model = Model(inputs=[head_input,body_input], outputs=[dense])
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['acc'])
print(model.summary())
````

### Compile and fit the model

````python
# Reshaping the features 
# For fitting input shape should be of 2 dim so we reshape feature from shape(x,20x20) to shape(x,400)
x_tr = x_train.reshape(x_train.shape[0],400)
x_tr_h = x_heading_train.reshape(x_heading_train.shape[0],400)
x_v = x_val.reshape(x_val.shape[0],400)
x_v_h = x_heading_val.reshape(x_heading_val.shape[0],400)
````

````python
print(x_train.shape,x_heading_train.shape,x_val.shape,x_heading_val.shape)
print(x_tr.shape,x_tr_h.shape,x_v.shape,x_v_h.shape)
````

````python
# Fitting the model
model.fit([x_tr[:10000],x_tr_h[:10000]], [y_train[:10000]], validation_data=([x_v,x_v_h], [y_val]), epochs=3, batch_size=512)
````

````python
#Accuracy of the model for train data 10000 samples
# As it takes time we took 10000 samples only with just 2 epochs.The accuracy will increase if the epochs are increased
model.evaluate([x_tr[:10000],x_tr_h[:10000]],y_train[:10000],batch_size=256)
````

## Build the same model with attention layers included for better performance (Optional)

````python
from keras.layers import Layer
import keras.backend as K
````

````python
class attention(Layer):
    def __init__(self,**kwargs):
        super(attention,self).__init__(**kwargs)

    def build(self,input_shape):
        self.W=self.add_weight(name="att_weight",shape=(input_shape[-1],1),initializer="normal")
        self.b=self.add_weight(name="att_bias",shape=(input_shape[1],1),initializer="zeros")        
        super(attention, self).build(input_shape)

    def call(self,x):
        et=K.squeeze(K.tanh(K.dot(x,self.W)+self.b),axis=-1) # score
        at=K.softmax(et) # alignment
        at=K.expand_dims(at,axis=-1)
        output=x*at # context vector
        return K.sum(output,axis=1)

    def compute_output_shape(self,input_shape):
        return (input_shape[0],input_shape[-1])

    def get_config(self):
        return super(attention,self).get_config()
````
